```python
import numpy as np
import matplotlib.pyplot as plt
import random

#Función para generar mi vector de valores aleatorios con una peso de probabilidad asignado
def random_vector (P,longitud2 ):  
    N= np.cumsum(np.random.choice([0, 1], size=longitud2, p=[1 - P, P]))
    N[0]=0
    return N

#Inserto los parámetros de mi problema
lam=3
repeticiones=10000
t=np.linspace(0,100,10000)
longitudT=len(t)
dt=t[1]-t[0]
P=lam*dt

#Matriz en la que iré guardando mis vectores de numeros aleatorios
M=np.zeros((repeticiones,len(t)))

#Matrices en las que guardaré los valores acumulados hasta cada tiempo 
R0=np.zeros(repeticiones)
R25=np.zeros(repeticiones)
R50=np.zeros(repeticiones)
R75=np.zeros(repeticiones)
R100=np.zeros(repeticiones)

for i in range(repeticiones):
    n=random_vector(P,longitudT)
    M[i,:]=n
    R0[i]=n[0]
    R25[i]=n[2500]
    R50[i]=n[5000]
    R75[i]=n[7500]
    R100[i]=n[9999]
    plt.plot(t,n)
plt.xlabel("t")
plt.title("")
plt.ylabel("N(t)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

#Función para generar mi vector de valores aleatorios con una peso de probabilidad asignado
def random_vector (b,d,longitud2 ):  
    N= np.cumsum(np.random.choice([-1,0, 1], size=longitud2, p=[d,1-(b+d), b]))
    return N

#Inserto los parámetros de mi problema
B=3
D=1
repeticiones=10000
t=np.linspace(0,100,10000)
longitudT=len(t)
dt=t[1]-t[0]
b=B*dt
d=D*dt

#Matriz en la que iré guardando mis vectores de numeros aleatorios
M=np.zeros((10000,len(t)))

#Matrices en las que guardaré los valores acumulados hasta cada tiempo 
R0=np.zeros(repeticiones)
R25=np.zeros(repeticiones)
R50=np.zeros(repeticiones)
R75=np.zeros(repeticiones)
R100=np.zeros(repeticiones)

for i in range(repeticiones):
    n=random_vector (b,d,longitudT )
    M[i,:]=n
    R0[i]=n[0]
    R25[i]=n[2500]
    R50[i]=n[5000]
    R75[i]=n[7500]
    R100[i]=n[9999]
    plt.plot(t,n)
plt.xlabel("t")
plt.title("")
plt.ylabel("N(t)")
```

```python
#Haciendo los histogramas
plt.hist(R0, bins=np.linspace(0,350,100))
plt.hist(R25, bins=np.linspace(0,350,100))
plt.hist(R50, bins=np.linspace(0,350,100))
plt.hist(R75, bins=np.linspace(0,350,100))
plt.hist(R100, bins=np.linspace(0,350,100))

plt.legend(['t=0','t=25', 't=50', 't=75', 't=100'])
plt.xlabel("W(t)")
plt.title("")
plt.ylabel("P(x,t)")

plt.plot((np.mean(R0), np.mean(R0)), (0, 3000), 'k--')
plt.plot((np.mean(R25), np.mean(R25)), (0, 3000), 'k--')
plt.plot((np.mean(R50), np.mean(R50)), (0, 3000), 'k--')
plt.plot((np.mean(R75), np.mean(R75)), (0, 3000), 'k--')
plt.plot((np.mean(R100), np.mean(R100)), (0, 3000), 'k--')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def DNI(beta, S, I, N, dt):
    P = np.array([-1, 1, 0])
    Weights = [(beta * I * S / N) * dt, 1 - (beta * I * S / N) * dt]
    A = np.random.choice([1, 0], p=Weights)
    return A * P

def DNR(gamma, I, dt):
    P = np.array([0, -1, 1])
    Weights = [gamma * I * dt, 1 - gamma * I * dt]
    A = np.random.choice([1, 0], p=Weights)
    return A * P
def SIR():
    NP = 1000
    gamma = 0.2
    beta = 0.8

    t0, tf, dt = 0, 100, 0.001

    t3 = np.arange(t0, tf + dt, dt)

    sigma = np.zeros([3, len(t3)])

    sigma[:,0]=[NP * .8, NP * 0.2, 0]

    for i in range(len(t3) - 1):
        dn1 = DNI(beta, sigma[0, i], sigma[1, i], NP, dt)
        dn2 = DNR(gamma, sigma[1, i], dt)
        sigma[:, i + 1] = sigma[:,i]+dn1+dn2
    
  
    return t3, sigma
```

```python
reps = 10
S_list = np.zeros((reps,len(t3)))
I_list = np.zeros((reps,len(t3)))
R_list = np.zeros((reps,len(t3)))
for i in range(10):
    t3, sigma = SIR()
    S_list[i,:] = sigma[0, :]
    I_list[i,:] = sigma[1, :]
    R_list[i,:] = sigma[2, :]
```

```python
for i in range(reps):
    plt.plot(t3, S_list[i])
    plt.plot(t3, I_list[i])
    plt.plot(t3, R_list[i])
    plt.title("MODELO SIR")
    plt.xlabel('Tiempo')
    plt.ylabel('Valores de sigma')
    plt.legend(['S', 'I', 'R'])
```

```python
promedio_t = [list(item) for item in zip(*S_list)]
promedio_t2 = []
for item in promedio_t:
  promedio_t2.append(np.mean(item))
plt.subplot(1, 2, 1)
plt.plot(t3,promedio_t2)
plt.title("Promedio de los Suceptibles")

varianza_t = [list(item) for item in zip(*S_list)]
varianza_t2 = []
for item in promedio_t:
  varianza_t2.append(np.var(item))
plt.subplot(1, 2, 2)
plt.plot(t3,varianza_t2)
plt.title("Varianza de los Suceptibles")
```
